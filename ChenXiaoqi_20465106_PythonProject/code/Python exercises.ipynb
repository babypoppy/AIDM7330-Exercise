```python
import sys
from PIL import Image
from IPython.display import display
```

```python
figuresPath = '../figures/'
```

```python
imageName = 'hopper'
imageName2 = 'lena'
imageExt = '.ppm'

imageFullName = figuresPath + imageName + imageExt
imageFullName2 = figuresPath + imageName2 + imageExt
im = Image.open(imageFullName)
im2 = Image.open(imageFullName2)
```

```python
im.show('hopper')
im2.show('lena')
```

```python
display(im)
display(im2)
```

```python
#print(im1.format, im1.size, im1.mode)
#print(im2.format, im2.size, im2.mode)
```

```python
thumbName = imageName + '_thumb'
thumbName2 = imageName2 + '_thumb2'
thumbExt = '.jpg'
thumbFullName = thumbName + thumbExt
thumbFullName2 = thumbName2 + thumbExt

size = (128, 128)

thumb = im.copy()
thumb2 = im2.copy()
thumb2.thumbnail(size)

thumb.save(thumbFullName, format = 'JPEG')
thumb2.save(thumbFullName2, format = 'JPEG')
display(thumb, thumb2)
```

```python
xsize, ysize = im.size
xsizeNew = xsize * 3
ysizeNew = ysize * 3

newIm = Image.new(mode = 'RGB', size = (xsizeNew, ysizeNew), color = 0)
for i in range(3):
    for k in range(3):
        if i == k or i + k == 2:
            newIm.paste(thumb,(i*xsize, k*ysize))
        else:
            newIm.paste(thumb2,(i*xsize, k*ysize))
            
display(newIm)
```
